```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
d=load_iris()
```

```python
d
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(d.data,d.target,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
s=SVC(C=20,gamma='auto',kernel='linear')
```

```python
s.fit(xtrain,ytrain)
```

```python
s.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(C=10,gamma='auto',kernel='linear'),d.data,d.target,cv=5)
```

```python
cross_val_score(SVC(C=10,gamma='auto',kernel='rbf'),d.data,d.target,cv=5)
```

```python
#GridSearch
```

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5)
```

```python
clf.fit(d.data,d.target)
clf.cv_results_
```

```python
a=pd.DataFrame(clf.cv_results_)
```

```python
a
```

```python
p=a[['param_C','param_kernel','mean_test_score']]
```

```python
p
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
#RandomizedSearch
```

```python
from sklearn.model_selection import RandomizedSearchCV
clf=RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,n_iter=2)
```

```python
clf.fit(d.data,d.target)
clf.cv_results_
```

```python
b=pd.DataFrame(clf.cv_results_)
```

```python
b[['param_kernel','param_C','mean_test_score']]
```

```python
clf.best_score_
```
