```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
a=pd.read_csv('customers.csv')
```

```python
a
```

```python
a.info()
```

```python
d1=a.iloc[:11,::]
```

```python
d1
```

```python
d1.info()
```

```python
#lineplot
```

```python
plt.plot(d1['Spending_Category'],d1['Spending_Score'],'r*--',ms=10,mec='b',mfc='g')
plt.xlabel('Spending_Category')
plt.ylabel('Spending_Score')
plt.title('Spending_Category vs Spending_Score')
```

```python
#Scatterplot
```

```python
plt.scatter(d1['Age'],d1['Annual_Income'])
plt.xlabel('Age')
plt.ylabel('Annual_Income')
plt.title('Age vs Annual_Income')
```

```python
#pie graph
```

```python
d=d1.groupby('Profession').size()
```

```python
d.index
```

```python
d.values
```

```python
plt.pie(d.values,labels=d.index,autopct='%.2f%%')
```

```python
#bar graph
```

```python
d2.index
```

```python
#Countplot
```

```python
a['Profession'].value_counts().index
```

```python
sns.countplot(x=a['Profession'],order=a['Profession'].value_counts().index)
plt.xticks(rotation=30)
```
