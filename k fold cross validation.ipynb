```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
d=load_digits()
```

```python
dir(d)
```

```python
d.data.shape
```

```python
d.target.shape
```

```python
p=pd.DataFrame(d.data)
p
```

```python
p['target']=d.target
p
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(p.drop(['target'],axis='columns'),p['target'],test_size=0.2)
```

```python
x_train
```

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
lg.score(x_test,y_test)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_test,y_test)
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
sv=SVC()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)
```

```python
def getscore(model,xtrain,ytrain,xtest,ytest):
    model.fit(xtrain,ytrain)
    print(model.score(xtest,ytest))
```

```python
getscore(DecisionTreeClassifier(),x_train,y_train,x_test,y_test)
```

```python
getscore(SVC(),x_train,y_train,x_test,y_test)
```

```python
getscore(RandomForestClassifier(),x_train,y_train,x_test,y_test)
```

```python
scores_logistic = []
scores_svm = []
scores_rf = []
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)


for train_index, test_index in folds.split(d.data,d.target):
    X_train, X_test, y_train, y_test = d.data[train_index], d.data[test_index], d.target[train_index], d.target[test_index]
   
    
    
    scores_logistic.append(getscore(LogisticRegression(), X_train, y_train,X_test, y_test))  
    scores_svm.append(getscore(SVC(),X_train, y_train,X_test, y_test))
    scores_rf.append(getscore(RandomForestClassifier(n_estimators=40), X_train, y_train,X_test, y_test))
print(scores_logistic)
print(scores_svm)
print(scores_rf)
```
